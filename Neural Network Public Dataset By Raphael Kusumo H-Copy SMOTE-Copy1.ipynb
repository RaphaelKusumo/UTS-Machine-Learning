```python
conditions = [
    (df['target'] >= 1),
    (df['target'] <= 0)
        ]
values = ['Passed', 'Failed']
df['Score'] = np.select(conditions, values)
df.head()
```

```python
#df = df.drop([ 'G3', 'target'],axis='columns')
df = df.drop(['target'],axis='columns')
df.head()
```

```python
inputs = df.drop('Score',axis='columns')
target = df['Score']
inputs.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(inputs)
```

```python
from sklearn.model_selection import train_test_split # Import train_test_split function
pred_train, pred_test, tar_train, tar_test = train_test_split(X_scaled, target, test_size=0.25, random_state=0)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

classifier = MLPClassifier( alpha=0.1626916610133735, learning_rate_init=0.01,
    hidden_layer_sizes=(50,), 
    activation='relu', 
    solver='adam', 
    max_iter=2000, 
    random_state=0
)

print("\nSedang melatih Neural Network...")
classifier.fit(pred_train, tar_train)
print("Pelatihan selesai!")

#Prediksi data tampil semua
predictions = classifier.predict(pred_test)
predictions

print("\n--- Laporan Klasifikasi ---")
print(classification_report(tar_test, predictions))

print("\n--- Confusion Matrix ---")
# Menampilkan berapa yang benar dan berapa yang salah tebak
print(confusion_matrix(tar_test, predictions))

# MCC score 1 Decision tree tanpa bagging
from sklearn.metrics import matthews_corrcoef
mcc_score = matthews_corrcoef(tar_test, predictions)
print("Matthews Correlation Coefficient:", mcc_score)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler


# 3. KONFIGURASI GRID SEARCH
# Model dasar
mlp_test = MLPClassifier(max_iter=500, random_state=0)

# Parameter yang ingin diuji coba
parameter_space = {
    'hidden_layer_sizes': [(30,30), (30,15), (50,), (100,)], # Struktur layer
    'activation': ['tanh', 'relu'],                 # Fungsi aktivasi
    'solver': ['adam', 'lbfgs'],                    # Algoritma optimizer
    'alpha': [0.0001, 0.05],                        # Regularisasi
}

# ---------------------------------------------------------
# 4. EKSEKUSI TUNING
# ---------------------------------------------------------
print("\nSedang mencari parameter terbaik (Grid Search)...")

# Inisialisasi Grid Search
clf = GridSearchCV(mlp_test, parameter_space, n_jobs=-1, cv=3)

# Lakukan training dengan variabel baru (pred_train, tar_train)
clf.fit(pred_train, tar_train)

# ---------------------------------------------------------
# 5. HASIL DAN EVALUASI
# ---------------------------------------------------------
print("\n--- HASIL TUNING ---")
print("Parameter Terbaik:", clf.best_params_)
print(f"Skor Akurasi Terbaik (saat validasi): {clf.best_score_:.4f}")

# Mengambil model juara
best_model = clf.best_estimator_

# Evaluasi menggunakan data test (pred_test, tar_test)
accuracy = best_model.score(pred_test, tar_test)
print(f"\nAkurasi Final di Data Test: {accuracy:.4f}")
```

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split, RandomizedSearchCV 
from sklearn.preprocessing import StandardScaler
from scipy.stats import uniform # Import untuk distribusi angka

# 3. KONFIGURASI GRID SEARCH
# ---------------------------------------------------------
# Model dasar
mlp_test = MLPClassifier(max_iter=500, random_state=0)

# Parameter yang ingin diuji coba
# 4. Parameter Space (Versi Hybrid: List + Distribusi)
parameter_space = {
    # Tetap pakai List untuk pilihan pasti
    'hidden_layer_sizes': [(30,30), (30,15), (50,), (100,), (50, 25), (100, 50)], 
    'activation': ['tanh', 'relu'],                 
    'solver': ['adam', 'lbfgs'],                    
    
    # Pakai uniform agar dia mencari angka terbaik di rentang ini
    'alpha': uniform(loc=0.0001, scale=0.1) 
}

# 5. Jalankan Randomized Search
print("Sedang mencari parameter terbaik secara acak...")
# n_iter=20 artinya kita menyuruh komputer mencoba 20 kombinasi acak
rs = RandomizedSearchCV(mlp_test, parameter_space, n_iter=20, n_jobs=-1, cv=3, random_state=0)

rs.fit(pred_train, tar_train)

# 6. Hasil
print("\nParameter Terbaik:", rs.best_params_)
print(f"Akurasi Terbaik: {rs.best_score_:.4f}")

# Test final
print(f"Akurasi di Data Test: {rs.score(pred_test, tar_test):.4f}")
```

```python
model = MLPClassifier(max_iter=2000, random_state=0) # Max iter diperbesar agar lbfgs konvergen

param_dist = {
    # 1. Struktur: Cenderung lebih sederhana/kecil
    'hidden_layer_sizes': [(10,), (30,), (50,), (30,15), (50,30), (100,)],
    
    # 2. Aktivasi: Tanh sering bekerja bagus untuk data numerik ternormalisasi
    'activation': ['tanh', 'relu'],
    
    # 3. Solver: LBFGS sangat direkomendasikan untuk data < 1000 baris
    'solver': ['lbfgs', 'adam'],
    
    # 4. Alpha: Mencari nilai penalti untuk mencegah overfitting
    'alpha': uniform(0.0001, 0.5),
    
    # 5. Learning Rate (Hanya berpengaruh jika solver='adam')
    'learning_rate_init': [0.001, 0.01, 0.005]
}

# ---------------------------------------------------------
# 5. JALANKAN PENCARIAN (RANDOMIZED SEARCH)
# ---------------------------------------------------------
print("\nSedang mencari parameter terbaik...")
# n_iter=50: Mencoba 50 kombinasi acak (Cukup banyak untuk hasil akurat)
# cv=5: Cross Validation 5 kali lipat (Standar yang baik)
rs = RandomizedSearchCV(
    model, 
    param_distributions=param_dist, 
    n_iter=50, 
    n_jobs=-1, 
    cv=5, 
    random_state=42,
    verbose=1
)

rs.fit(pred_train, tar_train)

# ---------------------------------------------------------
# 6. HASIL TERBAIK
# ---------------------------------------------------------
print("\n" + "="*40)
print("HASIL TUNING PARAMETER TERBAIK")
print("="*40)
print(f"Akurasi Terbaik (saat Training/Validasi): {rs.best_score_:.2%}")
print("\nParameter Juara:")
for param, val in rs.best_params_.items():
    print(f"- {param}: {val}")

# ---------------------------------------------------------
# 7. TES AKHIR
# ---------------------------------------------------------
final_model = rs.best_estimator_
acc_test = final_model.score(pred_test, tar_test)
print(f"\nAkurasi pada Data Test (Data Asing): {acc_test:.2%}")

# Cek Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(tar_test, final_model.predict(pred_test))
print("\nConfusion Matrix:")
print(cm)
```
